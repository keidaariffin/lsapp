```
text_list = reader.readtext("image2.JPG", add_margin=0.55, width_ths=0.7, link_threshold=0.8, decoder='beamsearch', blocklist='=-', detail=0)
text_list
```

```
text_comb=",".join(str(i) for i in text_list)
text_comb
```
